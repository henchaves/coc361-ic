```python
import numpy as np

from tensorflow.keras.layers import (
    Dense,
    Dropout,
)
from tensorflow.keras.models import Sequential
from scikeras.wrappers import KerasRegressor
from sklearn.model_selection import GridSearchCV
from tensorflow.keras.callbacks import ReduceLROnPlateau, EarlyStopping
from sklearn.metrics import r2_score, mean_squared_error
import pickle
```

```python
X = np.load("../data/X_train_scaled.npy", allow_pickle=True)
y = np.load("../data/y_train_scaled.npy", allow_pickle=True)
```

```python
print("X shape:", X.shape)
print("y shape:", y.shape)
```

```python
def create_model(n_neurons, dropout_rate, input_shape=(X.shape[1], )):
    model = Sequential()
    model.add(Dense(n_neurons, activation="relu", input_shape=input_shape))
    if dropout_rate > 0:
        model.add(Dropout(dropout_rate))
    model.add(Dense(n_neurons, activation="relu"))
    if dropout_rate > 0:
        model.add(Dropout(dropout_rate))
    model.add(Dense(1, activation="linear"))
    model.compile(optimizer="adam", loss="mse")
    return model
```

```python
regressor = KerasRegressor(
    model=create_model,
    n_neurons=32,
    dropout_rate=0,
    batch_size=512,
    epochs=40,
)
```

```python
param_grid = {
    "n_neurons": [32, 64],
    "dropout_rate": [0, 0.3]
}
```

```python
grid_search = GridSearchCV(
    regressor,
    param_grid,
    scoring="neg_mean_squared_error",
    cv=10,
    verbose=10,
    n_jobs=10
)
```

```python
np.random.seed(2907)
grid_result = grid_search.fit(X, y)
```

```python
print("GridSearch results:")
grid_search.cv_results_
```

```python
with open("../grids/grid_search_neural_net_02.pkl", mode="wb") as f:
    pickle.dump(grid_search, f)
```

```python
with open("../grids/grid_result_neural_net_02.pkl", mode="wb") as f:
    pickle.dump(grid_result, f)
```

```python
best_regressor = grid_search.best_estimator_
```

```python
print("Best regressor", best_regressor)
```

```python
with open("../models/model_gradient_neural_net_02.pkl", mode="wb") as f:
    pickle.dump(best_regressor, f)
```

```python
y_hat = best_regressor.predict(X)
```

```python
r2 = r2_score(y, y_hat)
mse = mean_squared_error(y, y_hat)

print(f"R2 Score: {r2:.3f}")
print(f"MSE: {mse:.3f}")
```
